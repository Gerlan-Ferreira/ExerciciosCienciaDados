```python
# Autor: Gerlan Ferreira, Matheus Araújo, Gustavo Belmont.
# Dataset Iris Flowers


from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
flores = pd.read_csv("/content/drive/MyDrive/Dados Ciencia De Dados/IRIS-flowers.csv",encoding="utf-8")
flores.head(10)
```

```python
flores.tail(1000)
```

```python
X = flores.iloc[:, 2:6].values
X
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 2, init = 'random')
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
distance = kmeans.fit_transform(X)
distance
```

```python
labels = kmeans.labels_
labels
```

```python
from sklearn.cluster import KMeans
wcss = []
i = 0 
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'random')
    kmeans.fit(X)
    print (i,kmeans.inertia_)
    wcss.append(kmeans.inertia_)  
plt.plot(range(1, 11), wcss)
plt.title('O Metodo Elbow')
plt.xlabel('Numero de Clusters')
plt.ylabel('WSS')
plt.show()
```

```python
import matplotlib.pyplot as plt
 
plt.scatter(X[:, 1], X[:,2], s = 100, c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 2], s = 300, c = 'red',label = 'Centroids')
plt.title('Flowers Clusters and Centroids')
plt.xlabel('species')
plt.ylabel('pétalas')
plt.legend('0 é setosa, 1 é virtinosa e 2 é virginica')
 
plt.show()
```
